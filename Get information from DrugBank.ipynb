```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
data = pd.read_excel("drug_id.xlsx")
```

```python
drug_ids = data["Drug"]
drug_ids = set(drug_ids)
```

```python
# Function to fetch drug name and description from DrugBank
def fetch_drug_details(drug_id):
    url = f"https://go.drugbank.com/drugs/{drug_id}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extract drug name
    drug_name_tag = soup.find("h1", class_="drug-name")
    drug_name = drug_name_tag.get_text(strip=True) if drug_name_tag else "N/A"
    
    # Extract drug description
    description_tag = soup.find("meta", {"name": "description"})
    description = description_tag["content"].strip() if description_tag else "N/A"
    
    return {"Drug ID": drug_id, "Drug Name": drug_name, "Description": description}

# Initialize an empty list to store results
results = []

# Process each DrugBank ID
for drug_id in drug_ids:
    print(f"Processing Drug ID: {drug_id}")
    try:
        drug_details = fetch_drug_details(drug_id)
        results.append(drug_details)
        print(f"Found details for Drug ID {drug_id}.")
    except Exception as e:
        print(f"Error processing Drug ID {drug_id}: {e}")
    time.sleep(2)  # Be polite to the server by adding a delay

# Convert results to a pandas DataFrame
df = pd.DataFrame(results)
```
