```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
gene_id = pd.read_excel("gene_id.xlsx")
```

```python
gene_ids = gene_id["gene_id"]
gene_ids = set(gene_ids)
gene_ids = list(gene_ids)
```

```python
# Convert to a list of strings
gene_ids = list(map(str, gene_ids))

# Print the result
print(gene_ids)
```

```python
# Function to fetch UniProtKB/Swiss-Prot entries for a gene ID
def fetch_uniprot_entries(gene_id):
    url = f"https://www.ncbi.nlm.nih.gov/gene/?term={gene_id}"
    response = requests.get(url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    
    entries = []
    for link in soup.find_all('a'):
        if "UniProtKB/Swiss-Prot" in link.get_text():
            entries.append(link.get_text().strip())
    return entries

# Initialize an empty list to store results
results = []
```

```python
# Process each gene ID
for gene_id in gene_ids:
    print(f"Processing Gene ID: {gene_id}")
    try:
        uniprot_entries = fetch_uniprot_entries(gene_id)
        for entry in uniprot_entries:
            results.append({"Gene ID": gene_id, "UniProt ID": entry})
        print(f"Found {len(uniprot_entries)} UniProt entries for Gene ID {gene_id}.")
    except Exception as e:
        print(f"Error processing Gene ID {gene_id}: {e}")
    time.sleep(1)  # Be polite to the server by adding a delay

# Convert results to a pandas DataFrame
df = pd.DataFrame(results)
```

```python
df.to_excel("gene_id_to_Uniprot.xlsx")
```
