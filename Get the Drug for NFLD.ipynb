```python
gene_list = data["List_gene"]
```

```python
# Misalkan gene_list adalah daftar string dengan format '[gene1, gene2, ...]'
# Contoh: gene_list = ['[geneA, geneB]', '[geneC, geneD]']

# Tempat untuk menyimpan hasil selected drugs
all_selected_drugs = []

# Loop untuk setiap gene_list
for i in range(len(gene_list)):
    gene_list_str = gene_list[i]
    # Bersihkan string dan pecah menjadi daftar
    genes = gene_list_str.strip('[]').replace(' ', '').split(',')
    print(genes)

    # Filter rows berdasarkan daftar gen
    subgraph = geneVsDrug.loc[geneVsDrug.index.isin(genes)]

    # Jumlah minimum yang dibutuhkan
    threshold = len(genes) / 2

    # Hitung jumlah pada setiap kolom
    column_sums = subgraph.sum(axis=0)

    # Filter kolom berdasarkan threshold
    selected_drugs = column_sums[column_sums >= threshold]

    # Simpan hasil untuk daftar gen ini
    all_selected_drugs.append(selected_drugs)

    # Hasil
    print("List drug above the threshold %d ******** in the cluster: %d" % (threshold, i))
    print(selected_drugs)
    print("-" * 50)

# Setelah semua iterasi, all_selected_drugs berisi hasil untuk setiap gene_list
```
